```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "deliveries.csv"
df = pd.read_csv(file_path)
plt.style.use("seaborn-v0_8-darkgrid")
```

```python
## Total Runs Scored Per Over
runs_per_over = df.groupby("over")["total_runs"].sum()

plt.figure(figsize=(12, 6))
sns.barplot(x=runs_per_over.index, y=runs_per_over.values, palette="Reds")
plt.xlabel("Over", fontsize=12)
plt.ylabel("Total Runs", fontsize=12)
plt.title("Total Runs Scored Per Over", fontsize=14)
plt.xticks(range(1, 21))
plt.show()
```

```python
# Display the top 10 players
print("Top 10 Players by Total Runs:")
print(runs_per_player.head(10))
```

```python
## Average Runs Per Over
average_runs_per_over = df.groupby("over")["total_runs"].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=average_runs_per_over.index, y=average_runs_per_over.values, marker="o", color="red")
plt.xlabel("Over", fontsize=12)
plt.ylabel("Average Runs", fontsize=12)
plt.title("Average Runs Per Over", fontsize=14)
plt.xticks(range(1, 21))
plt.show()

##Total Runs per Player
runs_per_player = df.groupby("batter")["batsman_runs"].sum().reset_index().sort_values(by="batsman_runs", ascending=False)
```

```python
## Top 10 Batsmen by Total Runs
top_batsmen = df.groupby("batter")["batsman_runs"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette="Reds")
plt.xlabel("Total Runs", fontsize=12)
plt.ylabel("Batsman", fontsize=12)
plt.title("Top 10 Batsmen by Total Runs", fontsize=14)
plt.show()
```

```python
## Runs Scored by Each Team
team_runs = df.groupby("batting_team")["total_runs"].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(y=team_runs.index, x=team_runs.values, palette="Reds")
plt.xlabel("Total Runs", fontsize=12)
plt.ylabel("Team", fontsize=12)
plt.title("Total Runs Scored by Each Team", fontsize=14)
plt.show()
```

```python
##Dismissal Types Distribution
dismissal_types = df["dismissal_kind"].dropna().value_counts()

plt.figure(figsize=(8, 8))
plt.pie(dismissal_types, labels=dismissal_types.index, autopct="%1.1f%%", colors=sns.color_palette("Reds", len(dismissal_types)))
plt.title("Dismissal Types Distribution", fontsize=14)
plt.show()
```

```python
##Extra Runs Analysis
extra_runs = df.groupby("extras_type")["extra_runs"].sum().dropna()

plt.figure(figsize=(8, 6))
sns.barplot(x=extra_runs.index, y=extra_runs.values, palette="Reds")
plt.xlabel("Extra Type", fontsize=12)
plt.ylabel("Total Extra Runs", fontsize=12)
plt.title("Total Extra Runs by Type", fontsize=14)
plt.show()
```

```python
##Strike Rate of Players (Runs per 100 Balls)
df["balls_faced"] = df.groupby("batter")["batter"].transform("count")
df["strike_rate"] = (df["batsman_runs"] / df["balls_faced"]) * 100

strike_rates = df.groupby("batter")["strike_rate"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=strike_rates.values, y=strike_rates.index, palette="Reds")
plt.xlabel("Strike Rate", fontsize=12)
plt.ylabel("Batsman", fontsize=12)
plt.title("Top 10 Batsmen by Strike Rate", fontsize=14)
plt.show()
```

```python
##Total Runs Per Match
runs_per_match = df.groupby("match_id")["total_runs"].sum()

plt.figure(figsize=(12, 6))
sns.histplot(runs_per_match, bins=30, kde=True, color="red")
plt.xlabel("Total Runs in Match", fontsize=12)
plt.ylabel("Number of Matches", fontsize=12)
plt.title("Distribution of Total Runs Per Match", fontsize=14)
plt.show()
```

```python
##Most Successful Bowlers (Highest Wickets Taken)
top_bowlers = df[df["is_wicket"] == 1].groupby("bowler")["is_wicket"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette="Reds")
plt.xlabel("Wickets Taken", fontsize=12)
plt.ylabel("Bowler", fontsize=12)
plt.title("Top 10 Bowlers by Wickets Taken", fontsize=14)
plt.show()
```

```python
##Strike Rate of Top Batsmen
batsmen_stats = df.groupby("batter").agg({"batsman_runs": "sum", "ball": "count"})
batsmen_stats["strike_rate"] = (batsmen_stats["batsman_runs"] / batsmen_stats["ball"]) * 100
top_strike_rate = batsmen_stats.sort_values(by="strike_rate", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_strike_rate["strike_rate"], y=top_strike_rate.index, palette="Reds")
plt.xlabel("Strike Rate", fontsize=12)
plt.ylabel("Batsman", fontsize=12)
plt.title("Top 10 Batsmen by Strike Rate", fontsize=14)
plt.show()
```

```python
##Economy Rate of Top Bowlers
bowler_stats = df.groupby("bowler").agg({"total_runs": "sum", "ball": "count"})
bowler_stats["overs"] = bowler_stats["ball"] / 6
bowler_stats["economy_rate"] = bowler_stats["total_runs"] / bowler_stats["overs"]
top_economy_bowlers = bowler_stats.sort_values(by="economy_rate").head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_economy_bowlers["economy_rate"], y=top_economy_bowlers.index, palette="Reds")
plt.xlabel("Economy Rate", fontsize=12)
plt.ylabel("Bowler", fontsize=12)
plt.title("Top 10 Bowlers by Economy Rate", fontsize=14)
plt.show()
```

```python
##Team-wise Run Contribution Per Player
teamwise_contribution = df.groupby(["batting_team", "batter"])["batsman_runs"].sum().reset_index()
top_contributions = teamwise_contribution.groupby("batting_team").apply(lambda x: x.sort_values("batsman_runs", ascending=False).head(3))

plt.figure(figsize=(12, 6))
sns.barplot(x=top_contributions["batsman_runs"], y=top_contributions["batter"], hue=top_contributions["batting_team"], dodge=False, palette="Set1")
plt.xlabel("Total Runs", fontsize=12)
plt.ylabel("Batsman", fontsize=12)
plt.title("Top Run Contributors Per Team", fontsize=14)
plt.legend(title="Team")
plt.show()
```

```python
##Most Common Partnerships (Top Batting Pairs)
batting_pairs = df.groupby(["batter", "non_striker"])["batsman_runs"].sum().reset_index()
top_partnerships = batting_pairs.sort_values(by="batsman_runs", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_partnerships["batsman_runs"], y=top_partnerships["batter"] + " & " + top_partnerships["non_striker"], palette="Reds")
plt.xlabel("Total Runs", fontsize=12)
plt.ylabel("Batting Pair", fontsize=12)
plt.title("Top 10 Batting Partnerships by Runs", fontsize=14)
plt.show()
```

```python
##Boundary Analysis (4s and 6s Per Batsman)
boundaries = df[df["batsman_runs"].isin([4, 6])]
boundary_count = boundaries.groupby(["batter", "batsman_runs"])["batsman_runs"].count().unstack(fill_value=0)

# Top 10 Batsmen based on boundaries
boundary_count["Total Boundaries"] = boundary_count[4] + boundary_count[6]
top_boundary_batsmen = boundary_count.sort_values("Total Boundaries", ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
top_boundary_batsmen[[4, 6]].plot(kind="bar", stacked=True, figsize=(12, 6), colormap="Reds", edgecolor="black")
plt.xlabel("Batsman", fontsize=12)
plt.ylabel("Number of Boundaries", fontsize=12)
plt.title("Top 10 Batsmen by Boundaries (4s & 6s)", fontsize=14)
plt.xticks(rotation=45)
plt.legend(["Fours", "Sixes"])
plt.show()
```
