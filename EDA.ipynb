```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/IPL_Data.csv')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df['BattingAVG'] = df['BattingAVG'].fillna(0)
df['BowlingAVG'] = df['BowlingAVG'].fillna(0)
df['EconomyRate'] = df['EconomyRate'].fillna(0)
df['S/R'] = df['S/R'].fillna(0)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['RunsScored'], kde=True)
plt.title('Distribution of Runs Scored')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Team', order = df['Team'].value_counts().index)
plt.title('Number of Players per Team')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='RunsScored', y='BattingAVG')
plt.title('Runs Scored vs. Batting Average')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Team', y='ValueinCR')
plt.title('Value of Players by Team')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.pairplot(df[['ValueinCR', 'RunsScored', 'Wickets', 'Age']])
plt.show()
```
