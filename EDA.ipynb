```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df = pd.read_csv("/content/matches (2).csv")
```

```python
print(df)
```

```python
# Set style
sns.set_style("whitegrid")
```

```python
# Check missing values
print("Missing Values:\n", df.isnull().sum())
```

```python
# Summary statistics
print("\nSummary Statistics:\n", df.describe())
```

```python
# Distribution of match results
plt.figure(figsize=(8, 5))
sns.countplot(x="result", data=df, palette="coolwarm")
plt.title("Distribution of Match Results")
plt.xlabel("Result Type")
plt.ylabel("Count")
plt.show()
```

```python
# Toss decision impact
plt.figure(figsize=(8, 5))
sns.countplot(x="toss_decision", data=df, palette="pastel")
plt.title("Toss Decision Distribution")
plt.xlabel("Toss Decision")
plt.ylabel("Count")
plt.show()
```

```python
# Top players
top_players = df["player_of_match"].value_counts().head(10)
plt.figure(figsize=(10, 5))
top_players.plot(kind="bar", color="skyblue")
plt.title("Top 10 Players with Most 'Player of the Match' Awards")
plt.xlabel("Player")
plt.ylabel("Awards Count")
plt.xticks(rotation=45)
plt.show()
```
