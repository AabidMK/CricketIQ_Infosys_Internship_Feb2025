```python
# seasons which had most no of matches and less no of matches
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='season',data=matches_data)
plt.show()
```

##### 2013 has more matches and 2009 has less matches #####

```python
# which team won more no of matches and less no of matches 
```

```python
plt.figure(figsize=(10,20))
data= matches_data.winner.value_counts()
sns.barplot(x=data,y=data.index)
plt.show()
```

##### mumbai indians won most of the matches and rising pune supergaints won less no of matches #####

```python
#man of the match
```

```python
top_players=matches_data.player_of_match.value_counts()[:10]
```

```python
top_players
```

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.set_ylim([0,40])
ax.set_ylabel("count")
ax.set_title("Man of the match winners")
top_players.plot.bar()
sns.barplot(x=top_players.index,y=top_players)
plt.show()
```

##### we can see mostly AB de Villiers is the man of the match for won matches #####
