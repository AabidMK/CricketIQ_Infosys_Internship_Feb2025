```python
# Total Runs Scored Per Over
over_runs = deliveries_df.groupby("over")["total_runs"].sum()
over_runs.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.xlabel("Over")
plt.ylabel("Total Runs")
plt.title("Total Runs Scored Per Over")
plt.show()
```

```python
# Dismissal Types Distribution
dismissal_counts = deliveries_df["dismissal_kind"].value_counts()
dismissal_counts.plot(kind='bar', figsize=(10, 5), color='coral')
plt.xlabel("Dismissal Type")
plt.ylabel("Count")
plt.title("Dismissal Types Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
# Total Extra Runs Conceded by Teams
extra_runs = deliveries_df.groupby("bowling_team")["extra_runs"].sum().sort_values(ascending=False)
extra_runs.plot(kind='bar', figsize=(12, 6), color='purple')
plt.xlabel("Team")
plt.ylabel("Total Extra Runs")
plt.title("Total Extra Runs Conceded by Teams")
plt.xticks(rotation=90)
plt.show()
```

```python
# Average Runs Per Over
average_runs_per_over = deliveries_df.groupby("over")["total_runs"].mean()
average_runs_per_over.plot(kind='line', marker='o', figsize=(10, 5), color='green')
plt.xlabel("Over")
plt.ylabel("Average Runs")
plt.title("Average Runs Per Over")
plt.show()
```

```python
# Distribution of Runs Per Ball
runs_per_ball = deliveries_df["batsman_runs"].value_counts()
runs_per_ball.plot(kind='bar', figsize=(10, 5), color='orange')
plt.xlabel("Runs Per Ball")
plt.ylabel("Count")
plt.title("Distribution of Runs Per Ball")
plt.show()
```

```python
# Top 10 Batsmen by Strike Rate
deliveries_df['balls_faced'] = deliveries_df.groupby("batter")["ball"].transform('count')
batsman_stats = deliveries_df.groupby("batter")["batsman_runs"].sum()
batsman_stats = batsman_stats / deliveries_df.groupby("batter")["balls_faced"].max() * 100
top_batsmen = batsman_stats.sort_values(ascending=False).head(10)
top_batsmen.plot(kind='bar', figsize=(10, 5), color='teal')
plt.xlabel("Batsman")
plt.ylabel("Strike Rate")
plt.title("Top 10 Batsmen by Strike Rate")
plt.show()
```

```python
# Most Sixes Hit by Players
sixes = deliveries_df[deliveries_df["batsman_runs"] == 6].groupby("batter").size().sort_values(ascending=False).head(10)
sixes.plot(kind='bar', figsize=(10, 5), color='red')
plt.xlabel("Batsman")
plt.ylabel("Number of Sixes")
plt.title("Most Sixes Hit by Players")
plt.show()
```

```python
# Dot Ball Percentage by Bowler
dot_balls = deliveries_df[deliveries_df["total_runs"] == 0].groupby("bowler").size()
total_balls = deliveries_df.groupby("bowler").size()
dot_ball_percentage = (dot_balls / total_balls * 100).dropna().sort_values(ascending=False).head(10)
dot_ball_percentage.plot(kind='bar', figsize=(10, 5), color='blue')
plt.xlabel("Bowler")
plt.ylabel("Dot Ball Percentage")
plt.title("Top 10 Bowlers with Highest Dot Ball Percentage")
plt.show()
```

```python
# Top 10 Bowlers Conceding Most Boundaries
boundaries = deliveries_df[deliveries_df["batsman_runs"] >= 4].groupby("bowler").size().sort_values(ascending=False).head(10)
boundaries.plot(kind='bar', figsize=(10, 5), color='darkred')
plt.xlabel("Bowler")
plt.ylabel("Boundaries Conceded")
plt.title("Top 10 Bowlers Conceding Most Boundaries")
plt.show()
```

```python
# Wicket Distribution Across Overs
wickets_per_over = deliveries_df[deliveries_df["is_wicket"] == 1].groupby("over").size()
wickets_per_over.plot(kind='bar', figsize=(12, 6), color='darkgreen')
plt.xlabel("Over")
plt.ylabel("Wickets")
plt.title("Wicket Distribution Across Overs")
plt.show()
```

```python
# Total Runs Scored in Powerplay and Death Over
powerplay_runs = deliveries_df[deliveries_df["over"] <= 6]["total_runs"].sum()
death_over_runs = deliveries_df[deliveries_df["over"] >= 16]["total_runs"].sum()
print("Total Runs in Powerplay:", powerplay_runs)
print("Total Runs in Death Overs:", death_over_runs)
```
