```python
import matplotlib.pyplot as plt
import plotly.express as pt
```

```python
#Total Runs Scored Per Over

Total_Scored_Per_Over = Deliveries_df.groupby('over')['total_runs'].sum().reset_index()
fig1 = pt.bar(Total_Scored_Per_Over, x = "over", y = "total_runs",title = "Total Runs Scored Per Over", color='total_runs',color_continuous_scale='burg')
fig1.show()
```

```python
#Dismissal Types Distribution
dismissal_counts = Deliveries_df["dismissal_kind"].value_counts().reset_index()
dismissal_counts.columns = ['dismissal_kind' ,'Count']
fig2 = pt.bar(dismissal_counts,x ="dismissal_kind",y="Count",title = "Dismissal Types Distribution",  color = 'Count',color_continuous_scale="twilight")
fig2.show()
```

```python
#Average Runs Per Over
Avg_runs = Deliveries_df.groupby('over')['total_runs'].mean().reset_index()
fig3 = pt.line(Avg_runs,x ='over',y='total_runs',title="Average Runs Per Over", markers = "o")
fig3.show()
```

```python
#Total Extra Runs Conceded By Teams
total_extra_runs = Deliveries_df.groupby('bowling_team')['extra_runs'].sum().reset_index().sort_values(by='extra_runs',ascending=False)
fig4 = pt.bar(total_extra_runs,x ='bowling_team',y='extra_runs',title="Total Extra Runs Conceded By Teams", color = 'extra_runs',color_continuous_scale='thermal')
fig4.show()

```

```python
#Distribution of Runs Per Ball(1s,2s,4s,6s)
Runs_Per_Ball = Deliveries_df.groupby('ball')['batsman_runs'].sum().reset_index()
fig5 = pt.bar(Runs_Per_Ball,x ='ball',y='batsman_runs',title="Distribution of Runs Per Ball(1s,2s,4s,6s)", color = 'batsman_runs',color_continuous_scale='algae')
fig5.show()
```

```python
#Top 10 Batsmen Hit By Strike Rate
batsmen_hit_strike = Deliveries_df.groupby("batter").agg(balls_faced=("ball","sum"),total_runs_scored=("batsman_runs","sum")).reset_index()
batsmen_hit_strike["Strike Rate"] = (batsmen_hit_strike["total_runs_scored"]/batsmen_hit_strike["balls_faced"])*100
ten_batsmen_strike = batsmen_hit_strike[batsmen_hit_strike["balls_faced"] >= 100].sort_values(by="Strike Rate", ascending=False).head(10)
fig6 = pt.bar(ten_batsmen_strike,x ='Strike Rate',y='batter',title="Top 10 Batsmen Hit By Strike Rate", color = 'Strike Rate',color_continuous_scale='delta')
fig6.show()
ten_batsmen_strike
#strike rate = (total runs/balls faced)*100
```

```python
#Most Sixes Hit By Player
Sixes = Deliveries_df[Deliveries_df["batsman_runs"]==6].groupby("batter").sum().reset_index()
fig7 = pt.bar(Sixes,x ='batter',y='batsman_runs',title="Most Sixes Hit By Player", color = 'batter')
fig7.show()
Sixes = Deliveries_df[Deliveries_df["batsman_runs"]==6].groupby("batter").sum().reset_index().sort_values(by="batsman_runs",ascending=False).head(15)
fig8 = pt.bar(Sixes,x ='batsman_runs',y='batter',title="Most Sixes Hit By Player", color = 'batsman_runs',color_continuous_scale='turbo')
fig8.show()
```

```python
#Dot Ball Percentage By Baller
Dot_Balls = Deliveries_df[Deliveries_df["total_runs"]==0].groupby("bowler").size()
Tot_Balls = Deliveries_df.groupby("bowler").size()
Dot_Ball_Percent = (Dot_Balls / Tot_Balls *100).sort_values(ascending=False).reset_index()
Dot_Ball_Percent.columns = ["bowler", "dot_ball_percentage"]
fig9 = pt.bar(Dot_Ball_Percent ,x ='bowler',y='dot_ball_percentage',title="Dot Ball Percentage By Baller", color = 'bowler',color_continuous_scale='turbo')
fig9.show()
#Top 10
Dot_Ball_Percent = (Dot_Balls / Tot_Balls *100).sort_values(ascending=False).reset_index().head(10)
Dot_Ball_Percent.columns = ["bowler", "dot_ball_percentage"]
fig10 = pt.bar(Dot_Ball_Percent ,x ='bowler',y='dot_ball_percentage',title="Dot Ball Percentage By Baller", color = 'dot_ball_percentage',color_continuous_scale='darkmint')
fig10.show()
```

```python
#Top 10 Bowler Conceding Most Boundaries
Bowler = Deliveries_df[Deliveries_df["batsman_runs"]>=4].groupby("bowler").size().sort_values(ascending=False).reset_index().head(10)
Bowler.columns = ["bowler", "Conceding_Boundaries"]                                                                                         
fig11 = pt.bar(Bowler ,x ='bowler',y='Conceding_Boundaries',title="Top 10 Bowler Conceding Most Boundaries", color = 'Conceding_Boundaries',color_continuous_scale='inferno')
fig11.show()
```

```python
#Wicket Distribution Across Overs
Wicket_Distribution = Deliveries_df.groupby('over')['is_wicket'].sum().reset_index()
Wicket_Distribution .columns = ["over", "Wickets"] 
fig12 = pt.bar(Wicket_Distribution,x ='over',y='Wickets',title="Wicket Distribution Across Overs", color = 'over',color_continuous_scale='blues')
fig12.show()
```
