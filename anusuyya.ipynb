```python
import pandas as pd

def preprocess_data():
    # Load dataset
    df = pd.read_csv('deliveries.csv')

    # Shape of dataset
    print("Shape of dataset:", df.shape)

    # Check for missing values
    print("\nMissing values:")
    print(df.isnull().sum())

    # Data types of columns
    print("\nData types:")
    print(df.dtypes)

    # Summary statistics
    print("\nSummary statistics:")
    print(df.describe())

    # Check for duplicate rows
    print("\nDuplicate rows:", df.duplicated().sum())

    return df

# Example usage
df = preprocess_data()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('deliveries.csv')

# Total Runs Scored Per Over
def plot_total_runs_per_over():
    over_runs = df.groupby('over')['total_runs'].sum().reset_index()
    plt.figure(figsize=(10, 5))
    sns.barplot(x='over', y='total_runs', data=over_runs, palette='Blues')
    plt.title('Total Runs Scored Per Over')
    plt.xlabel('Over')
    plt.ylabel('Total Runs')
    plt.show()

# Dismissal Types Distribution
def plot_dismissal_types():
    plt.figure(figsize=(10, 5))
    df['dismissal_kind'].dropna().value_counts().plot(kind='bar', color='skyblue')
    plt.title('Dismissal Types Distribution')
    plt.xlabel('Dismissal Type')
    plt.ylabel('Count')
    plt.show()

# Total Extra Runs Conceded by Teams
def plot_extra_runs_by_team():
    team_extras = df.groupby('bowling_team')['extra_runs'].sum().reset_index()
    plt.figure(figsize=(10, 5))
    sns.barplot(x='bowling_team', y='extra_runs', data=team_extras, palette='coolwarm')
    plt.xticks(rotation=90)
    plt.title('Total Extra Runs Conceded by Teams')
    plt.ylabel('Extra Runs')
    plt.xlabel('Teams')
    plt.show()

# Average Runs Per Over
def plot_avg_runs_per_over():
    over_avg = df.groupby('over')['total_runs'].mean().reset_index()
    plt.figure(figsize=(10, 5))
    sns.lineplot(x='over', y='total_runs', data=over_avg, marker='o')
    plt.title('Average Runs Per Over')
    plt.xlabel('Over')
    plt.ylabel('Average Runs')
    plt.show()

# Distribution of Runs Per Ball (1s, 2s, 4s, 6s)
def plot_run_distribution():
    run_types = df['batsman_runs'].value_counts().sort_index()
    plt.figure(figsize=(10, 5))
    run_types.plot(kind='bar', color=['blue', 'green', 'red', 'purple'])
    plt.title('Distribution of Runs Per Ball')
    plt.xlabel('Runs per Ball')
    plt.ylabel('Count')
    plt.show()

# Top 10 Batsmen by Strike Rate
def top_10_batsmen_strike_rate():
    batsmen_stats = df.groupby('batter').agg({'batsman_runs': 'sum', 'ball': 'count'})
    batsmen_stats['strike_rate'] = (batsmen_stats['batsman_runs'] / batsmen_stats['ball']) * 100
    top_batsmen = batsmen_stats.nlargest(10, 'strike_rate').reset_index()
    plt.figure(figsize=(10, 5))
    sns.barplot(x='batter', y='strike_rate', data=top_batsmen, palette='coolwarm')
    plt.xticks(rotation=90)
    plt.title('Top 10 Batsmen by Strike Rate')
    plt.show()

# Most Sixes Hit by Players
def most_sixes():
    sixes = df[df['batsman_runs'] == 6].groupby('batter').size().reset_index(name='sixes')
    top_six_hitters = sixes.nlargest(10, 'sixes')
    plt.figure(figsize=(10, 5))
    sns.barplot(x='batter', y='sixes', data=top_six_hitters, palette='rocket')
    plt.xticks(rotation=90)
    plt.title('Most Sixes Hit by Players')
    plt.show()

# Dot Ball Percentage by Bowler
def dot_ball_percentage():
    bowlers = df.groupby('bowler').agg({'ball': 'count', 'total_runs': lambda x: (x == 0).sum()})
    bowlers['dot_percentage'] = (bowlers['total_runs'] / bowlers['ball']) * 100
    plt.figure(figsize=(10, 5))
    sns.histplot(bowlers['dot_percentage'], bins=20, kde=True)
    plt.title('Dot Ball Percentage by Bowler')
    plt.xlabel('Dot Ball Percentage')
    plt.show()

# Top 10 Bowlers Conceding Most Boundaries
def top_10_bowlers_most_boundaries():
    boundaries = df[df['batsman_runs'] >= 4].groupby('bowler').size().reset_index(name='boundaries')
    top_bowlers = boundaries.nlargest(10, 'boundaries')
    plt.figure(figsize=(10, 5))
    sns.barplot(x='bowler', y='boundaries', data=top_bowlers, palette='magma')
    plt.xticks(rotation=90)
    plt.title('Top 10 Bowlers Conceding Most Boundaries')
    plt.show()

# Wicket Distribution Across Overs
def wicket_distribution():
    wickets = df[df['is_wicket'] == 1].groupby('over').size().reset_index(name='wickets')
    plt.figure(figsize=(10, 5))
    sns.barplot(x='over', y='wickets', data=wickets, palette='Set1')
    plt.title('Wicket Distribution Across Overs')
    plt.xlabel('Over')
    plt.ylabel('Wickets')
    plt.show()

# Total Runs Scored in Powerplay
def powerplay_runs():
    powerplay_runs = df[df['over'] < 6]['total_runs'].sum()
    print(f'Total Runs Scored in Powerplay: {powerplay_runs}')

# Total Runs Scored in Death Over
def death_over_runs():
    death_runs = df[df['over'] >= 16]['total_runs'].sum()
    print(f'Total Runs Scored in Death Overs: {death_runs}')

# Top 10 Most Economical Bowlers
def top_10_economical_bowlers():
    bowler_stats = df.groupby('bowler').agg({'total_runs': 'sum', 'over': 'count'})
    bowler_stats['economy'] = bowler_stats['total_runs'] / bowler_stats['over']
    top_bowlers = bowler_stats.nsmallest(10, 'economy').reset_index()
    plt.figure(figsize=(10, 5))
    sns.barplot(x='bowler', y='economy', data=top_bowlers, palette='coolwarm')
    plt.xticks(rotation=90)
    plt.title('Top 10 Most Economical Bowlers')
    plt.show()

# Call visualization functions
plot_total_runs_per_over()
plot_dismissal_types()
plot_extra_runs_by_team()
plot_avg_runs_per_over()
plot_run_distribution()
top_10_batsmen_strike_rate()
most_sixes()
dot_ball_percentage()
top_10_bowlers_most_boundaries()
wicket_distribution()
powerplay_runs()
death_over_runs()
top_10_economical_bowlers()
```
