```python
import plotly.express as px
```

# **Distribution of Total Runs per Match**

```python
match_wise_runs = deliveries.groupby('match_id')['total_runs'].sum().reset_index()
fig = px.histogram(match_wise_runs, x="total_runs", nbins=20, title="Distribution of Total Runs per Match")
fig.show()
```

# **Dismissal Kind Distribution Across Overs**

```python
dismissal_over_heatmap = deliveries[deliveries['player_dismissed'] != 'not dismissed'].pivot_table(index='over', columns='dismissal_kind', values='player_dismissed', aggfunc='count', fill_value=0)
fig = px.imshow(dismissal_over_heatmap, labels=dict(x="Dismissal Kind", y="Over", color="Dismissal Count"),
                 x=dismissal_over_heatmap.columns, y=dismissal_over_heatmap.index,
                 title='Dismissal Kind Distribution Across Overs')
fig.show()
```

# **Top Batter-Bowler Pairs by Total Runs**

```python
batsman_vs_bowler = deliveries.groupby(['bowler', 'batter'])['total_runs'].sum().reset_index()
top_pairs = batsman_vs_bowler.sort_values(by=['total_runs'], ascending=False).head(10)
fig = px.bar(top_pairs, x='batter', y='total_runs', color='bowler',
             title='Top Batter-Bowler Pairs by Total Runs',
             labels={'total_runs': 'Total Runs', 'batter': 'Batsman', 'bowler': 'Bowler'})
fig.show()
```

# **Distribution of Sixes per Over**

```python
sixes_per_over = deliveries[deliveries['batsman_runs'] == 6].groupby('over')['batsman_runs'].count()
fig = px.pie(sixes_per_over, values='batsman_runs', names=sixes_per_over.index,
             title="Distribution of Sixes per Over")
fig.show()
```

# **Inning with More Runs**

```python
total_runs_per_inning = deliveries.groupby(['match_id', 'inning'])['total_runs'].sum().reset_index()
inning_comparison = total_runs_per_inning.pivot_table(index='match_id', columns='inning', values='total_runs')
inning1_wins = len(inning_comparison[inning_comparison[1] > inning_comparison[2]])
inning2_wins = len(inning_comparison[inning_comparison[1] < inning_comparison[2]])
labels = ['Inning 1 Wins', 'Inning 2 Wins']
values = [inning1_wins, inning2_wins]
fig = px.pie(values=values, names=labels, hole=0.5, title='Inning with More Runs')
fig.show()
```

# **Wickets per Ball across all Overs**

```python
wickets_per_ball = deliveries[deliveries['player_dismissed'] != 'not dismissed'].groupby('ball')['player_dismissed'].count()
fig = px.bar(wickets_per_ball, x=wickets_per_ball.index, y='player_dismissed',
             title='Wickets per Ball across all Overs',
             labels={'player_dismissed': 'Number of Wickets', 'ball': 'Ball'})
fig.show()
```

# **Batting Team Performance (1s, 2s, 4s, 6s)**

```python
import plotly.graph_objects as go
runs_counts = deliveries.groupby('batting_team')['batsman_runs'].value_counts().unstack(fill_value=0)
runs_counts = runs_counts[[1, 2, 4, 6]]
runs_counts['total_runs'] = runs_counts.sum(axis=1)
runs_counts = runs_counts.sort_values('total_runs', ascending=False)
runs_counts = runs_counts.drop('total_runs',axis=1)
fig = go.Figure()

for run_type in [1, 2, 4, 6]:
  fig.add_trace(go.Bar(x=runs_counts.index,y=runs_counts[run_type],name=str(run_type)))
fig.update_layout(title='Batting Team Performance (1s, 2s, 4s, 6s)',xaxis_title='Team Name',yaxis_title='Count',barmode='stack')
fig.show()
```

# **Top 10 Bowlers with Most Overs of Zero Runs**

```python
zero_runs_overs = deliveries[deliveries['total_runs'] == 0].groupby('bowler')['total_runs'].count()
top_10_bowlers_zero_runs = zero_runs_overs.sort_values(ascending=False).head(10)
fig = px.bar(top_10_bowlers_zero_runs,
             x=top_10_bowlers_zero_runs.index,
             y=top_10_bowlers_zero_runs.values,
             labels={'x':'Bowler', 'y':'Count of Overs with 0 Runs'},
             title='Top 10 Bowlers with Most Overs of Zero Runs')
fig.show()
```
