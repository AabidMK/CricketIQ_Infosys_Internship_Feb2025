```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('IPL_Data.csv')
```

```python
def clean_age(age_str):
    if isinstance(age_str, str):
        parts = age_str.split(', ')
        years = int(parts[0].split(' ')[0])
        return years
    return None
```

```python
df['Age'] = df['Age'].apply(clean_age)
```

```python
df['ValueinCR'] = pd.to_numeric(df['ValueinCR'], errors='coerce')
```

```python
df['RunsScored'] = pd.to_numeric(df['RunsScored'], errors='coerce')
```

```python
df['Wickets'] = pd.to_numeric(df['Wickets'], errors='coerce')
```

```python
df = df.fillna(0)
```

```python
top_10_strike_rate = df.sort_values(by='BattingS/R', ascending=False).head(10)
print("Top 10 Batsmen by Strike Rate:\n", top_10_strike_rate[['Full Name', 'BattingS/R']])
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_strike_rate, x='Full Name', y='BattingS/R')
plt.title('Top 10 Batsmen by Strike Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
most_sixes = df.sort_values(by='6s', ascending=False).head(10)
print("Players with Most Sixes:\n", most_sixes[['Full Name', '6s']])
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=most_sixes, x='Full Name', y='6s')
plt.title('Players with Most Sixes')
plt.xticks(rotation=45)
plt.show()
```

```python
most_economical_bowlers = df.sort_values(by='EconomyRate', ascending=True).head(10)
print("Top 10 Most Economical Bowlers:\n", most_economical_bowlers[['Full Name', 'EconomyRate']])
```
