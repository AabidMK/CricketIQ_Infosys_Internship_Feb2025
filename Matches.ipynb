```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Handling missing null values
df.isnull()
```

```python
df[df['winner'].isnull()]
```

```python
df['winner'].fillna("No Result", inplace=True)
df['result_margin'].fillna(0, inplace=True)
df['target_runs'].fillna(0, inplace=True)
df['target_overs'].fillna(0, inplace=True)

# Drop 'method' column since it's mostly missing
df.drop(columns=['method'], inplace=True)

# Verify that missing values are handled
df.isnull().sum()
```

```python
df.describe()
```

```python
df['winner'].value_counts()
```

```python
most_successful_team = df['winner'].value_counts().idxmax()
print("Most successful team:", most_successful_team)
```

```python
# Calculate total matches played by each team (as winner or loser)
total_matches_played = df['winner'].value_counts() + df['winner'].value_counts()
win_percentage = (df['winner'].value_counts() / total_matches_played) * 100

# Remove "No Result" from data
win_percentage = win_percentage.drop("No Result", errors='ignore')

# Plot win percentage
plt.figure(figsize=(12, 6))
sns.barplot(x=win_percentage.index, y=win_percentage.values, palette="viridis")
plt.xticks(rotation=90)
plt.xlabel("Teams")
plt.ylabel("Win Percentage (%)")
plt.title("Teams with Highest Win Percentage")
plt.show()
```

```python
top_teams=df['winner'].value_counts().head(5)
#bar graph for top 10 teams
plt.figure(figsize=(10,3))
sns.barplot(x=top_teams.values, y=top_teams.index, palette="viridis")
plt.xlabel("Teams")
plt.ylabel("Top 5 teams")
```
