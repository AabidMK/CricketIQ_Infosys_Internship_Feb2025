```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better visuals
sns.set_style("whitegrid")

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="season", order=df["season"].value_counts().index, palette="viridis")
plt.xticks(rotation=45)
plt.title("Number of Matches per Season")
plt.xlabel("Season")
plt.ylabel("Match Count")
plt.show()
```

```python
# Count the number of wins per team
top_teams = df["winner"].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_teams.index, y=top_teams.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Teams with Most Wins")
plt.xlabel("Team")
plt.ylabel("Win Count")
plt.show()
```
