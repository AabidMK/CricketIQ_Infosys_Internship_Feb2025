```python
# Ensure the relevant columns exist (modify if needed)
if 'Overs' in df.columns and 'RunsScored' in df.columns:
    # Group by 'over' and sum the total runs
    runs_per_over = df.groupby('Overs')['RunsScored'].sum()

    # Plot
    plt.figure(figsize=(10, 5))
    plt.bar(runs_per_over.index, runs_per_over.values, color='blue')
    # Labels and title
    plt.xlabel("Over", fontsize=12)
    plt.ylabel("Total Runs", fontsize=12)
    plt.title("Total Runs Scored Per Over in IPL Matches", fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Columns 'over' and 'total_runs' not found in the dataset.")
```

```python
df1 = pd.read_csv('deliveries.csv')
```

```python
# Ensure the 'dismissal_kind' column exists
if 'dismissal_kind' in df1.columns:
    # Count occurrences of each dismissal type
    dismissal_counts = df1['dismissal_kind'].value_counts()

    # Plot
    plt.figure(figsize=(10, 5))
    plt.bar(dismissal_counts.index, dismissal_counts.values, color='crimson')

    # Labels and title
    plt.xlabel("Dismissal Type", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.title("Dismissal Types Distribution in IPL Matches", fontsize=14)
    plt.xticks(rotation=45)  # Rotate labels for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Column 'dismissal_kind' not found in the dataset.")
```

```python
# Ensure the required columns exist
if 'Team' in df.columns and 'RunsConceded' in df.columns:
    # Group by bowling team and sum extra runs
    extra_runs_by_team = df.groupby('Team')['RunsConceded'].sum()

    # Plot
    plt.figure(figsize=(12, 6))
    plt.bar(extra_runs_by_team.index, extra_runs_by_team.values, color='purple')

    # Labels and title
    plt.xlabel("Team", fontsize=12)
    plt.ylabel("Total Extra Runs", fontsize=12)
    plt.title("Total Extra Runs Conceded by Teams in IPL Matches", fontsize=14)
    plt.xticks(rotation=45)  # Rotate labels for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Columns 'bowling_team' and 'extra_runs' not found in the dataset.")
```

```python
# Ensure the required columns exist
if 'Overs' in df.columns and 'RunsScored' in df.columns:
    # Group by over and calculate the average runs per over
    avg_runs_per_over = df.groupby('Overs')['RunsScored'].mean()

    # Plot
    plt.figure(figsize=(25, 5))
    plt.plot(avg_runs_per_over.index, avg_runs_per_over.values, marker='o', linestyle='-', color='green')

    # Labels and title
    plt.xlabel("Over", fontsize=12)
    plt.ylabel("Average Runs", fontsize=12)
    plt.title("Average Runs Per Over in IPL Matches", fontsize=14)
    plt.xticks(sorted(avg_runs_per_over.index.dropna().astype(int))) # Ensure integer x-axis labels
    plt.xticks(rotation=90)
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Columns 'over' and 'total_runs' not found in the dataset.")
```

```python
# Ensure the required column exists
if 'batsman_runs' in df1.columns:
    # Filter only valid scoring shots (1, 2, 4, 6)
    scoring_runs = df1[df1['batsman_runs'].isin([1, 2, 4, 6])]

    # Count occurrences of each run type
    run_distribution = scoring_runs['batsman_runs'].value_counts().sort_index()

    # Plot
    plt.figure(figsize=(8, 5))
    plt.bar(run_distribution.index, run_distribution.values, color=['blue', 'orange', 'red', 'green'])

    # Labels and title
    plt.xlabel("Runs Per Ball", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.title("Distribution of Runs Per Ball (1s, 2s, 4s, 6s)", fontsize=14)
    plt.xticks(run_distribution.index)  # Ensure only 1, 2, 4, and 6 appear on x-axis
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Column 'batsman_runs' not found in the dataset.")
```

```python
# Ensure required columns exist
if {'Name', 'BattingS/R'}.issubset(df.columns):
    # Convert 'BattingS/R' to numeric (handle errors if any non-numeric values exist)
    df['BattingS/R'] = pd.to_numeric(df['BattingS/R'], errors='coerce')

    # Get the Top 10 Batsmen by Strike Rate
    top_10_batsmen = df.nlargest(10, 'BattingS/R')

    # Plot
    plt.figure(figsize=(12, 6))
    plt.bar(top_10_batsmen['Name'], top_10_batsmen['BattingS/R'], color='purple')

    # Labels and title
    plt.xlabel("Batsman", fontsize=12)
    plt.ylabel("Strike Rate", fontsize=12)
    plt.title("Top 10 Batsmen by Strike Rate in IPL", fontsize=14)
    plt.xticks(rotation=45)  # Rotate labels for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Columns 'Name' and 'BattingS/R' not found in the dataset.")
```

```python
# Ensure required columns exist
if {'Name', '6s'}.issubset(df.columns):
    # Convert '6s' to numeric (handle errors if any non-numeric values exist)
    df['6s'] = pd.to_numeric(df['6s'], errors='coerce')

    # Get the Top 10 players with the most sixes
    top_10_six_hitters = df.groupby('Name')['6s'].sum().nlargest(10)

    # Plot
    plt.figure(figsize=(12, 6))
    plt.bar(top_10_six_hitters.index, top_10_six_hitters.values, color='orange')

    # Labels and title
    plt.xlabel("Batsman", fontsize=12)
    plt.ylabel("Total Sixes", fontsize=12)
    plt.title("Top 10 Players with Most Sixes in IPL", fontsize=14)
    plt.xticks(rotation=45)  # Rotate labels for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Columns 'Name' and '6s' not found in the dataset.")
```

```python
# Ensure required columns exist
if {'bowler', 'total_runs'}.issubset(df1.columns):
    # Consider dot balls as deliveries where 'Runs Given' is 0
    df1['Dot Ball'] = df1['total_runs'] == 0

    # Group by bowler to calculate total balls and dot balls
    bowler_stats = df1.groupby('bowler').agg(
        total_balls=('bowler', 'count'),
        dot_balls=('Dot Ball', 'sum')  # Counting dot balls
    )

    # Calculate Dot Ball Percentage
    bowler_stats['Dot Ball %'] = (bowler_stats['dot_balls'] / bowler_stats['total_balls']) * 100

    # Get the Top 10 bowlers with the highest Dot Ball %
    top_10_bowlers = bowler_stats.nlargest(10, 'Dot Ball %')

    # Plot
    plt.figure(figsize=(12, 6))
    plt.bar(top_10_bowlers.index, top_10_bowlers['Dot Ball %'], color='blue')

    # Labels and title
    plt.xlabel("Bowler", fontsize=12)
    plt.ylabel("Dot Ball Percentage", fontsize=12)
    plt.title("Top 10 Bowlers by Dot Ball Percentage in IPL", fontsize=14)
    plt.xticks(rotation=45)  # Rotate labels for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Columns 'Bowler' and 'Runs Given' not found in the dataset.")
```

```python
# Ensure required columns exist
if {'Name', '4s', '6s'}.issubset(df.columns):
    # Convert '4s' and '6s' to numeric (handle errors if any non-numeric values exist)
    df['4s'] = pd.to_numeric(df['4s'], errors='coerce')
    df['6s'] = pd.to_numeric(df['6s'], errors='coerce')

    # Calculate total boundaries conceded (fours + sixes)
    df['Total Boundaries'] = df['4s'] + df['6s']

    # Group by bowler to sum the total boundaries conceded
    bowler_boundaries = df.groupby('Name')['Total Boundaries'].sum()

    # Get the Top 10 bowlers who conceded the most boundaries
    top_10_bowlers = bowler_boundaries.nlargest(10)

    # Plot
    plt.figure(figsize=(12, 6))
    plt.bar(top_10_bowlers.index, top_10_bowlers.values, color='red')

    # Labels and title
    plt.xlabel("Bowler", fontsize=12)
    plt.ylabel("Total Boundaries Conceded", fontsize=12)
    plt.title("Top 10 Bowlers Conceding Most Boundaries in IPL", fontsize=14)
    plt.xticks(rotation=45)  # Rotate labels for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Columns 'Bowler', '4s', and '6s' not found in the dataset.")
```

```python
#using deliveries.csv
# Ensure required columns exist
if {'over', 'is_wicket'}.issubset(df1.columns):
    # Convert 'Wicket' column to numeric (assuming 1 for wicket, NaN/0 otherwise)
    df1['is_wicket'] = pd.to_numeric(df1['is_wicket'], errors='coerce')

    # Count the number of wickets per over
    wickets_per_over = df1.groupby('over')['is_wicket'].sum()

    # Plot
    plt.figure(figsize=(25, 6))
    plt.bar(wickets_per_over.index, wickets_per_over.values, color='green')

    # Labels and title
    plt.xlabel("Overs", fontsize=12)
    plt.ylabel("Number of Wickets", fontsize=12)
    plt.title("Wicket Distribution Across Overs in IPL", fontsize=14)
    plt.xticks(range(int(min(wickets_per_over.index)), int(max(wickets_per_over.index)) + 1))
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()
else:
    print("Columns 'Over' and 'Wicket' not found in the dataset.")
```

```python
# Ensure required columns exist
if {'Overs', 'RunsScored'}.issubset(df.columns):
    # Filter for powerplay overs (1-6)
    powerplay_runs = df[(df['Overs'] >= 1) & (df['Overs'] <= 6)]

    # Calculate total runs scored in the powerplay
    total_powerplay_runs = powerplay_runs['RunsScored'].sum()

    print(f"Total Runs Scored in Powerplay (Overs 1-6): {total_powerplay_runs}")
else:
    print("Columns 'over' and 'total_runs' not found in the dataset.")
```

```python
# Ensure required columns exist
if {'Overs', 'RunsScored'}.issubset(df.columns):
    # Filter for death overs (16-20)
    death_over_runs = df[(df['Overs'] >= 16) & (df['Overs'] <= 20)]

    # Calculate total runs scored in the death overs
    total_death_over_runs = death_over_runs['RunsScored'].sum()

    print(f"Total Runs Scored in Death Overs (Overs 16-20): {total_death_over_runs}")
else:
    print("Columns 'over' and 'total_runs' not found in the dataset.")
```

```python
# Ensure required columns exist
if {'Name', 'EconomyRate', 'Overs'}.issubset(df.columns):
    # Convert to numeric (handle errors)
    df['EconomyRate'] = pd.to_numeric(df['EconomyRate'], errors='coerce')
    df['Overs'] = pd.to_numeric(df['Overs'], errors='coerce')

    # Filter out bowlers with less than 10 overs bowled
    filtered_bowlers = df[df['Overs'] >= 10]

    # Get the Top 10 Most Economical Bowlers (sorted by lowest economy rate)
    top_10_economical = filtered_bowlers.nsmallest(10, 'EconomyRate')

    # Plot
    plt.figure(figsize=(12, 6))
    plt.barh(top_10_economical['Name'], top_10_economical['EconomyRate'], color='yellow')
    plt.xlabel("Economy Rate", fontsize=12)
    plt.ylabel("Bowler", fontsize=12)
    plt.title("Top 10 Most Economical Bowlers in IPL", fontsize=14)
    plt.gca().invert_yaxis()  # Invert to show lowest economy first
    plt.grid(axis='x', linestyle='--', alpha=0.7)

    # Show the plot
    plt.show()

else:
    print("Columns 'bowler', 'economy_rate', and 'overs_bowled' not found in the dataset.")
```
