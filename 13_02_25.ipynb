```python
df = pd.read_csv('/content/matches.csv')

pivot_table = df.pivot_table(values='result_margin', index='team1', columns='team2', aggfunc='mean')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt='.1f', cmap='coolwarm', linewidths=0.5)
plt.title('Average Winning Margin Between Teams')
plt.xlabel('Team 2')
plt.ylabel('Team 1')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

matches = pd.read_csv('/content/matches.csv')
deliveries = pd.read_csv('/content/deliveries.csv')

print("Matches Columns:", matches.columns)
print("Deliveries Columns:", deliveries.columns)

runs_per_over = deliveries.groupby('over')['total_runs'].sum()
plt.figure(figsize=(10, 5))
sns.barplot(x=runs_per_over.index, y=runs_per_over.values, hue=runs_per_over.index, legend=False, palette='viridis')
plt.title('Total Runs Scored Per Over')
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.show()

plt.figure(figsize=(10, 5))
sns.countplot(y=deliveries['dismissal_kind'].dropna(), hue=deliveries['dismissal_kind'].dropna(), legend=False, palette='coolwarm')
plt.title('Dismissal Types Distribution')
plt.xlabel('Count')
plt.ylabel('Dismissal Type')
plt.show()

extra_runs = deliveries.groupby('bowling_team')['extra_runs'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 5))
sns.barplot(x=extra_runs.index, y=extra_runs.values, hue=extra_runs.index, legend=False, palette='rocket')
plt.title('Total Extra Runs Conceded by Teams')
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Total Extra Runs')
plt.show()

avg_runs_per_over = deliveries.groupby('over')['total_runs'].mean()
plt.figure(figsize=(10, 5))
sns.lineplot(x=avg_runs_per_over.index, y=avg_runs_per_over.values, marker='o', color='b')
plt.title('Average Runs Per Over')
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.show()

runs_per_ball = deliveries['batsman_runs'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
sns.barplot(x=runs_per_ball.index, y=runs_per_ball.values, hue=runs_per_ball.index, legend=False, palette='magma')
plt.title('Distribution of Runs Per Ball')
plt.xlabel('Runs Scored on a Ball')
plt.ylabel('Frequency')
plt.show()

if 'batsman' in deliveries.columns:
    batsman_stats = deliveries.groupby('batsman').agg({'batsman_runs': 'sum', 'ball': 'count'})
    batsman_stats['strike_rate'] = (batsman_stats['batsman_runs'] / batsman_stats['ball']) * 100
    top_batsmen = batsman_stats.nlargest(10, 'strike_rate')

    plt.figure(figsize=(12, 5))
    sns.barplot(y=top_batsmen.index, x=top_batsmen['strike_rate'], hue=top_batsmen.index, legend=False, palette='plasma')
    plt.title('Top 10 Batsmen by Strike Rate')
    plt.xlabel('Strike Rate')
    plt.ylabel('Batsman')
    plt.show()

if 'batsman' in deliveries.columns:
    sixes = deliveries[deliveries['batsman_runs'] == 6].groupby('batsman')['batsman_runs'].count().nlargest(10)
    plt.figure(figsize=(12, 5))
    sns.barplot(y=sixes.index, x=sixes.values, hue=sixes.index, legend=False, palette='cividis')
    plt.title('Most Sixes Hit by Players')
    plt.xlabel('Total Sixes')
    plt.ylabel('Batsman')
    plt.show()

dot_balls = deliveries[deliveries['total_runs'] == 0].groupby('bowler')['ball'].count()
total_balls = deliveries.groupby('bowler')['ball'].count()
dot_ball_percentage = (dot_balls / total_balls * 100).dropna().nlargest(10)

plt.figure(figsize=(12, 5))
sns.barplot(y=dot_ball_percentage.index, x=dot_ball_percentage.values, hue=dot_ball_percentage.index, legend=False, palette='coolwarm')
plt.title('Dot Ball Percentage by Bowler')
plt.xlabel('Dot Ball Percentage')
plt.ylabel('Bowler')
plt.show()

boundaries = deliveries[deliveries['batsman_runs'].isin([4, 6])].groupby('bowler')['batsman_runs'].count().nlargest(10)

plt.figure(figsize=(12, 5))
sns.barplot(y=boundaries.index, x=boundaries.values, hue=boundaries.index, legend=False, palette='inferno')
plt.title('Top 10 Bowlers Conceding Most Boundaries')
plt.xlabel('Total Boundaries Conceded')
plt.ylabel('Bowler')
plt.show()

wickets_per_over = deliveries[deliveries['dismissal_kind'].notna()].groupby('over')['dismissal_kind'].count()
plt.figure(figsize=(10, 5))
sns.barplot(x=wickets_per_over.index, y=wickets_per_over.values, hue=wickets_per_over.index, legend=False, palette='Blues')
plt.title('Wicket Distribution Across Overs')
plt.xlabel('Over')
plt.ylabel('Wickets')
plt.show()

powerplay_runs = deliveries[deliveries['over'] <= 6]['total_runs'].sum()
print(f'Total Runs Scored in Powerplay: {powerplay_runs}')

death_over_runs = deliveries[deliveries['over'] >= 16]['total_runs'].sum()
print(f'Total Runs Scored in Death Overs: {death_over_runs}')

economy_stats = deliveries.groupby('bowler').agg({'total_runs': 'sum', 'ball': 'count'})
economy_stats['economy'] = (economy_stats['total_runs'] / economy_stats['ball']) * 6
top_economical_bowlers = economy_stats.nsmallest(10, 'economy')

plt.figure(figsize=(12, 5))
sns.barplot(y=top_economical_bowlers.index, x=top_economical_bowlers['economy'], hue=top_economical_bowlers.index, legend=False, palette='Greens_r')
plt.title('Top 10 Most Economical Bowlers')
plt.xlabel('Economy Rate')
plt.ylabel('Bowler')
plt.show()
```
