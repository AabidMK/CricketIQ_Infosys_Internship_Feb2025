```python
import pandas as pd
import numpy as np
df=pd.read_csv("deliveries (1).csv")
df
```

Total Runs Scored Per Over

```python
runs_per_over = df.groupby("over")["total_runs"].sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.bar(runs_per_over.index, runs_per_over.values, color='violet')
plt.xlabel("over")
plt.ylabel("total_runs")
plt.title("Total Runs Scored Per Over")
plt.xticks(runs_per_over.index)
plt.show()
```

```python
sixes_by_players = df[df["batsman_runs"] == 6].groupby("batter")["batsman_runs"].count().nlargest(10)
```

```python
plt.figure(figsize=(12, 6))
sixes_by_players.plot(kind='bar', color='orange')
plt.xlabel("Player")
plt.ylabel("Number of Sixes")
plt.title("Top 10 Players with Most Sixes")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
boundaries_by_bowlers = df[df["batsman_runs"].isin([4, 6])].groupby("bowler")["batsman_runs"].count().nlargest(10)
```

```python
plt.figure(figsize=(12, 6))
boundaries_by_bowlers.plot(kind='bar', color='yellow')
plt.xlabel("Bowler")
plt.ylabel("Number of Boundaries Conceded")
plt.title("Top 10 Bowlers Conceding Most Boundaries")
plt.xticks(rotation=45)
plt.show()
```

```python
boundaries_by_bowlers = df[df["batsman_runs"].isin([4, 6])].groupby("bowler")["batsman_runs"].count().nlargest(10)
```

```python
plt.figure(figsize=(12, 6))
boundaries_by_bowlers.plot(kind='bar', color='red')
plt.xlabel("Bowler")
plt.ylabel("Number of Boundaries Conceded")
plt.title("Top 10 Bowlers Conceding Most Boundaries")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
runs_per_ball = df["batsman_runs"].value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
runs_per_ball.plot(kind='bar', color='teal')
plt.xlabel("Runs Per Ball")
plt.ylabel("Count")
plt.title("Distribution of Runs Per Ball (1s, 2s, 4s, 6s)")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
wickets_per_over = df[df["dismissal_kind"].notna()].groupby("over").size()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(wickets_per_over.index, wickets_per_over.values, color='brown')
plt.xlabel("Over Number")
plt.ylabel("Number of Wickets")
plt.title("Wicket Distribution Across Overs")
plt.xticks(wickets_per_over.index)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
