```python
#1.Total runs scored per over
```

```python
runs_per_over = df2.groupby(["match_id", "inning", "over"])["total_runs"].sum().reset_index()
```

```python
sample_match_id = runs_per_over["match_id"].iloc[0]
match_data = runs_per_over[runs_per_over["match_id"] == sample_match_id]
```

```python
plt.figure(figsize=(12, 6))
for inning in match_data["inning"].unique():
    inning_data = match_data[match_data["inning"] == inning]
    plt.plot(inning_data["over"], inning_data["total_runs"], marker="o", label=f"Inning {inning}")

plt.xlabel("Over")
plt.ylabel("Total Runs")
plt.title(f"Runs Scored Per Over for Match ID {sample_match_id}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#2.Dismissal types distribution
```

```python
dismissal_counts = df2["dismissal_kind"].value_counts()
```

```python
plt.figure(figsize=(12, 6))
dismissal_counts.plot(kind="bar", color="skyblue")
plt.xlabel("Dismissal Type")
plt.ylabel("Count")
plt.title("Distribution of Dismissal Types")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
#3.Average runs per over
```

```python
total_matches = df2["match_id"].nunique()
average_runs_per_over = runs_per_over.groupby("over")["total_runs"].sum() / total_matches
```

```python
plt.figure(figsize=(12, 6))
average_runs_per_over.plot(kind="bar", color="coral")
plt.xlabel("Over")
plt.ylabel("Average Runs")
plt.title("Average Runs Per Over Across All Matches")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.show()
```

```python
#4.Top 10 Batsmen by strike rate
```

```python
batsman_stats = df2.groupby("batter").agg({"batsman_runs": "sum", "ball": "count"}).reset_index()
batsman_stats["strike_rate"] = (batsman_stats["batsman_runs"] / batsman_stats["ball"]) * 100
```

```python
top_batsmen = batsman_stats.sort_values(by="strike_rate", ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_batsmen["batter"], top_batsmen["strike_rate"], color="purple")
plt.xlabel("Batsman")
plt.ylabel("Strike Rate")
plt.title("Top 10 Batsmen by Strike Rate")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
#5.Top 10 most economical bowlers
```

```python
bowler_stats = df2.groupby("bowler").agg({"total_runs": "sum", "ball": "count"}).reset_index()
bowler_stats["overs"] = bowler_stats["ball"] / 6
bowler_stats = bowler_stats[bowler_stats["overs"] > 0]
bowler_stats["economy_rate"] = bowler_stats["total_runs"] / bowler_stats["overs"]
```

```python
top_bowlers = bowler_stats.sort_values(by="economy_rate").head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_bowlers["bowler"], top_bowlers["economy_rate"], color="green")
plt.xlabel("Bowler")
plt.ylabel("Economy Rate")
plt.title("Top 10 Most Economical Bowlers")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
#6.top 10 bowlers conceding across overs
```

```python
bowler_over_runs = df2.groupby(["bowler", "over"])["total_runs"].sum().reset_index()
```

```python
top_conceding_bowlers = bowler_over_runs.groupby("bowler")["total_runs"].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
top_conceding_bowlers.plot(kind="bar", color="red")
plt.xlabel("Bowler")
plt.ylabel("Total Runs Conceded")
plt.title("Top 10 Bowlers Conceding the Most Runs Across Overs")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
#7.Dot ball percentage by bolwer
```

```python
top_dot_bowlers = (df2.assign(dot_ball=lambda x: x["total_runs"] == 0)
                     .groupby("bowler").agg(dot_balls=("dot_ball", "sum"), balls=("ball", "count"))
                     .assign(dot_ball_percentage=lambda x: (x["dot_balls"] / x["balls"]) * 100)
                     .nlargest(10, "dot_ball_percentage"))
```

```python
plt.figure(figsize=(10, 6))
plt.pie(top_dot_bowlers["dot_ball_percentage"],
    labels=top_dot_bowlers.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Blues_r(range(len(top_dot_bowlers))))
```
