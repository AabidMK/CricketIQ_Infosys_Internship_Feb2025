```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df['date'] = pd.to_datetime(df['date'])
df['date']
```

```python
plt.figure(figsize=(14,9))
```

```python
plt.title("No. of Matches/Season")
plt.ylabel("Total No. Of Matches")
plt.xlabel("Season")
sns.countplot(data=df, x = "season", order=sorted(df["season"].unique()),palette="Blues")
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle="-")
plt.show()
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.title("Tosses Win By Team")
sns.barplot(x=df["toss_winner"].value_counts().index,y=df["toss_winner"].value_counts().values,palette="Purples")
plt.xticks(rotation=90)
plt.xlabel("Toss Winning Team")
plt.ylabel("Total No. of Tosses")
plt.grid(axis='y',linestyle="-")
plt.show()
```

```python
plt.title("Toss Winning Decisions")
sns.countplot(data=df,x="toss_winner",hue="toss_decision",order=df["toss_winner"].value_counts().index, palette="Greys")
plt.xticks(rotation=90)
plt.xlabel("Toss Winning Team")
plt.ylabel("Decisons Counts")
plt.legend(title="Toss Decisions")
plt.show()
```

```python
plt.figure(figsize=(13,8))
plt.title("Match Winning Teams")
sns.barplot(y=df["winner"].value_counts().index,x=df["winner"].value_counts().values,palette="RdBu")
plt.xticks(rotation=90)
plt.xlabel("Total Wins")
plt.ylabel("Winning Teams")
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.title("Top 11 Players")
df["player_of_match"].unique()
players=df["player_of_match"].value_counts().head(11)
sns.barplot(x=players.values,y=players.index,palette="PiYG")
plt.xticks(rotation=90)
plt.ylabel("Players")
plt.xlabel("Counts")
plt.show()
```
